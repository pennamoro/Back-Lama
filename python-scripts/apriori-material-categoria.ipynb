```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("receitas-material-categoria.csv", header=None, skiprows=1, names=['materiais_concatenados'])
df['materiais_concatenados'] = df['materiais_concatenados'].str.split(', ')
new_df = pd.DataFrame(df['materiais_concatenados'].to_list())
result_df = pd.concat([df, new_df], axis=1)

result_df.replace(np.nan, 0, inplace=True)
result_df.drop(columns=['materiais_concatenados'], inplace=True)

df = result_df
df.head(10)
```

```python
def removeAllZeros(l):
    return list(filter(lambda x: x != 0, l))
```

```python
first_list = []
for index, row in df.iterrows():
    listOfMaterial = row.values.tolist()
    listOfMaterial = removeAllZeros(listOfMaterial)

    first_list.append(listOfMaterial)

list_all_material = [[item.split(' -')[0] for item in sublist] for sublist in first_list]

list_all_material[0:10]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(list_all_material).transform(list_all_material)
df = pd.DataFrame(te_ary, columns=te.columns_)

df.head(10)
```

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(df, min_support = 0.03, use_colnames= True)
frequent_itemsets.sort_values(by=['support'], ascending= False)
```

```python
from mlxtend.frequent_patterns import association_rules

rules = association_rules(frequent_itemsets, metric = 'confidence', min_threshold= 0.5)
sorted_rules = rules.sort_values(by=['confidence'], ascending= False).drop(['antecedent support', 'consequent support', 'leverage', 'conviction', 'support', 'zhangs_metric'], axis=1)
sorted_rules
```

```python
sorted_rules['antecedents'] = sorted_rules['antecedents'].apply(lambda x: list(x))
sorted_rules['consequents'] = sorted_rules['consequents'].apply(lambda x: list(x))

sorted_rules.to_csv("regras-material-categoria.csv", index=False)
```
